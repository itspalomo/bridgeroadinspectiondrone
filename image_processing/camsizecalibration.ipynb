```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline  
import seaborn as sns
```

```python
sns.set_theme(style="white", palette="pastel")
```

```python
vl_image = cv2.imread('camera.jpg')
ir_image = cv2.imread('infra_red.jpg')
print(f'Visible light info: \n', vl_image.shape, '\nIR Info: ', ir_image.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Original Images')
fig.set_figheight(5)
fig.set_figwidth(20)

ax1.imshow(vl_image)
ax1.set_title('Visible Light')

ax2.imshow(ir_image)
ax2.set_title('Infrared')
```
